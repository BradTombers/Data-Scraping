```python
from selenium import webdriver
from getpass import getpass
from selenium.webdriver.support.select import Select 
import selenium
import time
import csv
import collections

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome('/Users/b.tombers/Documents/chromedriverpath/chromedriver')


#team_ids = ['1610612738', '1610612751', '1610612766', '1610612741', '1610612739', '1610612742',
 #           '1610612743','1610612765','1610612744','1610612745','1610612754','1610612746','1610612747','1610612763'
  #          ,'1610612748','1610612749','1610612750','1610612740','1610612752','1610612760','1610612753',
   #         '1610612755','1610612756','1610612757','1610612758','1610612759','1610612761','1610612762','1610612737','1610612764']  

team_ids = ['1610612761','1610612762','1610612737','1610612764']  

url = 'http://stats.nba.com/team/'

for team in team_ids:
   # csv_file = open('nba_rosters' + team + '.csv', 'w')
    driver.get(url + team)
    time.sleep(3)
    
    # get team name
    team_name = ''
    team_name_list = driver.find_elements_by_xpath('/html/body/main/div[2]/div/div/div[2]/div/div/div[1]/div/div/div[2]/div[1]')
    for thing in team_name_list:
        team_name = team_name + thing.text
        print(team_name) 
    team_name_list = driver.find_elements_by_xpath('/html/body/main/div[2]/div/div/div[2]/div/div/div[1]/div/div/div[2]/div[2]')
    for thing in team_name_list:
        team_name = team_name + ' ' + thing.text
        print(team_name)
        
    obj = Select(driver.find_element_by_name("Season"))
    options = obj.options
    team_dict = collections.OrderedDict()
    csv_file = open('nba_rosters_' + team + '.csv', 'w')
    writer = csv.writer(csv_file)
    writer.writerow(['TEAM','TEAMID','SEASON', 'PLAYER', 'PLAYERID','NO','POS','HEIGHT','WEIGHT','BIRTHDATE','AGE','EXP','SCHOOL'])
    
    for option in options:        
        print('start of ', option.text) 
        obj.select_by_visible_text(option.text)
        time.sleep(2)
        players = driver.find_elements_by_xpath('/html/body/main/div[2]/div/div/div[3]/div/div/div[1]/nba-stat-table/div[2]/div[1]/table/tbody/tr')

        for player in players:  
            
            season = option.text
            player_name = player.find_element_by_xpath('.//td/a').text                
          #  print(player_name)
            player_id = str(player.find_element_by_xpath('.//td/a').get_attribute('href')).replace('https://stats.nba.com/player/','').replace('/','')
          #  print(player_id)
            no = player.find_element_by_xpath('.//td[2]').text
          #  print(no)
            pos = player.find_element_by_xpath('.//td[3]').text
          #  print(pos)
            height = player.find_element_by_xpath('.//td[4]').text
          #  print(height)
            weight = player.find_element_by_xpath('.//td[5]').text
          #  print(weight)
            birthdate = player.find_element_by_xpath('.//td[6]').text
          #  print(birthdate)
            age = player.find_element_by_xpath('.//td[7]').text
          #  print(age)
            exp = player.find_element_by_xpath('.//td[8]').text
          #  print(exp)
            school = player.find_element_by_xpath('.//td[9]').text
          #  print(school)

            team_dict['TEAM'] = team_name
            team_dict['TEAMID'] = team
            team_dict['SEASON'] = season
            team_dict['PLAYER'] = player_name
            team_dict['PLAYERID'] = player_id
            team_dict['NO'] = no
            team_dict['POS'] = pos
            team_dict['HEIGHT'] = height
            team_dict['WEIGHT'] = weight
            team_dict['BIRTHDATE'] = birthdate
            team_dict['AGE'] = age
            team_dict['EXP'] = exp
            team_dict['SCHOOL'] = school

            writer.writerow(team_dict.values())

        print('end of ', option.text)
       
    csv_file.close()
            
```
