```python
import urllib.request
import csv
from bs4 import BeautifulSoup
from selenium import webdriver  
from selenium.webdriver.common.keys import Keys  
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException


# Initiate the webdriver

chrome_options = webdriver.ChromeOptions()

driver = webdriver.Chrome('/Users/b.tombers/Documents/chromedriverpath/chromedriver')


# A lazy way to tell the program which team abbreviation/season combinations to skip since the GSIS site has malfunctiong pages for them

skip_teams = [['SD', '2017'], ['SL', '2016'], ['SL', '2017'], ['HST', '2000'], ['HST', '2001'], ['HST', '1999']]
skip_teams += [['LAC',str(x)] for x in range(1999,2017)] + [['LA',str(x)] for x in range(1999,2016)]


# Log in to site

driver.get('http://nflgsis.com/GameStatsLive/Auth/')

username = driver.find_element_by_id("Username")
password = driver.find_element_by_id("Password")

username.send_keys('espn')
password.send_keys('sports')

driver.find_element_by_id('btnLogin').click()

# Navigate to Legacy Reports page

driver.get('http://nflgsis.com/GameStatsLive/LegacyReports')

driver.find_element_by_xpath('//*[@id="content"]/nav/ul/li[2]/a').click()
driver.find_element_by_xpath('//*[@id="reportList"]/ul/li[2]/a').click()

final_dat = []

# Loop through the teams

for team in driver.find_elements_by_xpath('//*[@id="filterPanel"]/div/ul/li[3]/select/option'):

	team.click()

	team_abbr = team.get_attribute('value')

	team_name = team.text

	# Loop through seasons

	for season in driver.find_elements_by_xpath('//*[@id="filterPanel"]/div/ul/li[1]/select/option'):

		season.click()

		season_name = season.text

		print([team_abbr,season_name])

		# Skip teams if they're in the list

		if [team_abbr,season_name] in skip_teams:
			print('Skipping ' + team_name + ' ' + str(season_name))
			continue


		driver.switch_to.frame(driver.find_element_by_name('BodyData'))
		print('Working on ' + team_name + ' ' + str(season_name))

		soup = BeautifulSoup(driver.page_source,'html.parser')

		driver.switch_to.default_content()

		# Loops 

		for num in range(2,18):
			data_list = (team_name,str(season_name),str(num-1))

			try: 
				tbl = soup.find('table', attrs = {'id': 'hgrdPenaltiesByGame__ctl' + str(num) + '_ChildTemplate_GameKeyRel_grdPlayerPenalties'}).tbody
			except:
				continue
				
			for tr in tbl.find_all('tr'):

				row_data = []

				for item in data_list:
						row_data.append(item)

				for td in tr.find_all('td'):

					item = td.get_text()

					if isinstance(item,int):
						item = str(item)
					else:
						item = item

					row_data.append( item )

				final_dat.append(row_data)

driver.quit()

# Write to a csv

# Weird workaround for new Python3 error
new_line_in_bytes = '\n'
new_line_in_bytes = new_line_in_bytes.encode('utf-8')

with open('pst_pen_test.csv', 'wb') as file:
    for line in final_dat:
        if len(line) >= 4:
            dat = ','.join(line).encode('utf-8').strip()
            dat = dat + new_line_in_bytes
            file.write(dat)
            
```
